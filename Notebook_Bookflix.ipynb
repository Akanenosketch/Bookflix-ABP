# Cargar dataset en un DataFrame

Primero cargamos el dataset data.csv en un DataFrame de pandas para poder manipularlo y analizarlo de manera eficiente.
(Nota : Primero es necesario cargar manualmente el archivo al notebook)

Un DataFrame es una estructura tabular que permite realizar operaciones como filtrar, ordenar, agregar columnas o calcular estad√≠sticas.
Mostramos una vista previa de los datos para comprobar que se han cargado correctamente.

```python
import pandas as pd

df = pd.read_csv('data.csv')
df
```

# A√±adir un nuevo libro

En esta celda definimos un nuevo libro con toda su informaci√≥n: ISBN13, ISBN10, t√≠tulo, subt√≠tulo, autores, categor√≠a, imagen, descripci√≥n, a√±o de publicaci√≥n, valoraci√≥n media, n√∫mero de p√°ginas y n√∫mero de valoraciones.

Luego, utilizamos `pd.concat` para a√±adir este libro al DataFrame existente.  
Finalmente, mostramos las √∫ltimas filas del DataFrame para comprobar que el libro se ha a√±adido correctamente.

Esta operaci√≥n es √∫til para realizar un registro de nuevos libros en nuestro dataset sin sobrescribir los existentes.

```python
# Definir el nuevo libro
nuevo_libro = {
    "isbn13": 9789999999999,
    "isbn10": "9999999999",
    "title": "El Arte de Programar",
    "subtitle": "Edici√≥n Avanzada",
    "authors": "Donald Knuth",
    "categories": "Inform√°tica",
    "thumbnail": "https://ejemplo.com/arte_programar.jpg",
    "description": "Gu√≠a completa sobre estructuras y algoritmos.",
    "published_year": 2025,
    "average_rating": 4.9,
    "num_pages": 950,
    "ratings_count": 4200
}

# A√±adirlo al DataFrame
df = pd.concat([df, pd.DataFrame([nuevo_libro])], ignore_index=True)

# Comprobar que se a√±adi√≥
df.tail(2)
```

# Modificar informaci√≥n de un libro

En esta celda actualizamos los datos de un libro existente, buscando primero su ISBN13.  
Modificamos el t√≠tulo y la valoraci√≥n media (`average_rating`) para reflejar cambios en la edici√≥n o en las valoraciones.

Mostrar las columnas modificadas nos permite verificar que los cambios se aplicaron correctamente, sin afectar otros libros.

```python
# Ver el libro antes de modificarlo
df.loc[df["isbn13"] == 9789999999999, ["title", "average_rating"]]
```

```python
# Modificar campos
df.loc[df["isbn13"] == 9789999999999, "title"] = "Gilead (Edici√≥n Revisada)"
df.loc[df["isbn13"] == 9789999999999, "average_rating"] = 4.2

# Ver los cambios
df.loc[df["isbn13"] == 9789999999999, ["title", "average_rating"]]
```

# Filtrar libros por categor√≠a

Podemos seleccionar todos los libros de una categor√≠a espec√≠fica para analizarlos m√°s a fondo o visualizarlos por separado.  
En este ejemplo filtramos todos los libros cuya categor√≠a es "Inform√°tica".

Esto es √∫til para trabajar con subconjuntos de datos y realizar an√°lisis espec√≠ficos de cada tipo de libro.

```python
# Filtrar libros de la categor√≠a "Inform√°tica"
libros_informatica = df[df["categories"] == "Inform√°tica"]
libros_informatica
```

# Eliminar un libro

A veces necesitamos eliminar registros del dataset, por ejemplo si un libro se ha retirado o se ha a√±adido por error.

Primero mostramos cu√°ntos libros hay antes de la eliminaci√≥n.  
Luego filtramos el DataFrame para excluir el libro con el ISBN especificado.  
Finalmente, mostramos cu√°ntos libros quedan despu√©s de la eliminaci√≥n para comprobar que se ha borrado correctamente.

```python
# Ver cu√°ntos libros hay antes
print("Libros antes:", len(df))

# Eliminar el libro por ISBN
df = df[df["isbn13"] != 9789999999999]  # El ISBN del libro a√±adido antes

# Ver cu√°ntos quedan
print("Libros despu√©s:", len(df))
```

# Ordenar libros por valoraci√≥n media

Es √∫til ordenar los libros seg√∫n su valoraci√≥n media (`average_rating`) para identificar los m√°s recomendados.  
Podemos ordenar de mayor a menor para ver los libros mejor valorados o de menor a mayor para detectar los peor valorados.

La funci√≥n `sort_values()` permite especificar la columna por la que ordenar y si queremos orden ascendente o descendente.

```python
# Ordenar de mayor a menor valoraci√≥n
df_sorted = df.sort_values(by="average_rating", ascending=False)
df_sorted.head(10)  # Mostrar los 10 mejores valorados
```

# Contar libros por categor√≠a

A veces queremos conocer cu√°ntos libros tenemos en cada categor√≠a.  
Esto nos ayuda a entender la distribuci√≥n de nuestro cat√°logo y a identificar categor√≠as m√°s o menos representadas.  

Usamos `value_counts()` sobre la columna `categories`, que devuelve un recuento de cu√°ntas veces aparece cada categor√≠a en el DataFrame.

```python
# Contar libros por categor√≠a
df["categories"].value_counts()
```

# Estad√≠sticas descriptivas

Podemos obtener estad√≠sticas b√°sicas como media, m√≠nimo, m√°ximo y desviaci√≥n est√°ndar
de columnas num√©ricas como `num_pages` o `average_rating`.

```python
# Estad√≠sticas del n√∫mero de p√°ginas y valoraci√≥n media
df[["num_pages", "average_rating"]].describe()
```

# Buscar libros de un autor espec√≠fico

Podemos filtrar todos los libros escritos por un autor determinado.  
En este ejemplo buscamos libros de "George Orwell".  

Algunos valores de la columna `authors` pueden estar vac√≠os (`NaN`).  
Por eso usamos `na=False` en `str.contains()` para evitar errores y que las filas vac√≠as no se incluyan en el resultado.

```python
# Filtrar libros de J.K. Rowling, ignorando los NaN
df[df["authors"].str.contains("George Orwell", na=False)]
```
